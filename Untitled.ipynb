```python
with open('mobile.html','r',encoding ='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
import numpy as np
```

```python
soup.prettify()
```

```python
containers =(soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}))
```

```python
name = []
price = []
rating = []
sim = []
processor =[]
ram = []
battery = []
display =[]
camera=[]

for i in containers:
    name.append(i.find('h2').text)
    price.append(i.find('span',{'class':'price'}).text)
    try:
        rating.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        rating.append(np.nan)
    x=i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)


    try:
        ram.append(x[2].text)
    except:
        ram.append(np.nan)


    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)

     
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    

    try:
        camera.append(x[4].text)
    except:
        camera.append(np.nan)
```

```python
len(name)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'model':name,
    'price':price,
    'rating':rating,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera

})
```
